## **All important functions and methods:**
- pd.Series()
- pd.Dataframe()
- pd.read_csv()
  - df.head() - first 5 rows
  - df.tail() - last 5 rows
  - df.shape
  - df.ndim - number of dimensions
  - df.dtypes - data type of columns
  - df.describe() - stats about data  ---> `df['radius_se'].describe()`  # describe a specific column
  - `df['area_se'].value_counts()` - counts distinct values
    - `df['area_se'].value_counts(ascending=True) ` - asc
    - `df['area_se'].value_counts(dropna=False)`- shows NaN
    - `df2['area_se'].value_counts(normalize=True)` - gives percentage
    - `df2['area_se'].value_counts()[16.97]` - gives specific item
    - `df2.groupby('area_se').size() - count unique values` - size
    - `df2.groupby('area_se').count() - count unique values `- count
  - `df2['nothingnew'] = [True] * len(df2)` - adding a new column
  - `df2.insert(1, "new", [True]*len(df2), True)` # index, header, whats inside, allow duplicates  - adding new columns with insert
  - `df3 = df2.assign(Horse=[False]*len(df2))` - add columns by creating a new dataframe
  - `df2.drop("area_mean", axis=1)` - remove columns axis 1 (by default its 0)
  - `df2.drop("area_mean", axis=1, inplace=True)` - removes it permanently (inplace = True)
  - `df2.loc[2] `- grab row method 1
  - `df2.iloc[1]` - grab row method 2 - index location
  - `df2.loc[1, 'radius_mean']` - grab specific rows and columns
  - `df2.loc[[1, 3], ['area_worst', 'radius_mean']]` - grab subsets
