# LINEAR REGRESSION

import libraries

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
import matplotlib.pyplot as pyplot
import pickle
from matplotlib import style
```

read csv and choose correct columns - https://colab.research.google.com/drive/1A_6dBwCb35P__qWqtjkKduCyZ-TDyP48?authuser=1

```python
data = pd.read_csv('student-mat.csv', sep=';')
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
pyplot.scatter(data["G3"], data["studytime"])
pyplot.xlabel("Final Grade")
```

we want to predict G3 so X = data without G3, Y = only G3 data

```python
predict = "G3"
x = np.array(data.drop([predict], 1)) #drop deletes data, 1 is axis
y = np.array(data["G3"])
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.1)

# you always get 4 parameters - x_train, x_test, y_train, y_test as an output

# x train - parameteres which contains every other value except G3
# x test - parameteres of x_train but to test
# y_train - parameters which contains only G3, actual values of G3
# y_test - parameters of y_train but to test, actual values of G3

# test_size is the size of x_test value, it means that 20 % of the whole dataset is for testing
# x_train contains the values that are random, if you dont want to change it, use random_state=int

'''
best = 0
for _ in range(500):

  linear = linear_model.LinearRegression()
  linear.fit(x_train, y_train) # data to train a model, obviously contains only training data

  acc = linear.score(x_test, y_test) # to verify your accuracy

  if float(acc) > float(best):
    best = acc
    print(best)
    print(acc)
    with open("studentmodel.pickle", "wb") as f:
      pickle.dump(linear, f)
'''
```

1). X_train - This includes your all independent variables,these will be used to train the model, also as we have specified the test_size = 0.4, this means 60% of observations from your complete data will be used to train/fit the model and rest 40% will be used to test the model.

2). X_test - This is remaining 40% portion of the independent variables from the data which will not be used in the training phase and will be used to make predictions to test the accuracy of the model.

3). y_train - This is your dependent variable which needs to be predicted by this model, this includes category labels against your independent variables, we need to specify our dependent variable while training/fitting the model.

4). y_test - This data has category labels for your test data, these labels will be used to test the accuracy between actual and predicted categories

```python
pickle_in = open("studentmodel.pickle", "rb")
linear = pickle.load(pickle_in)

print("Co", linear.coef_) # coefficient - współczynnik a
print("intercept:", linear.intercept_) # intercept - współczynnik b

predictions = linear.predict(x_test)

for x in range(len(predictions)):
  print(predictions[x], x_test[x], y_test[x])
  if round(predictions[x]) == y_test[x]:
    print("correct")
pyplot.scatter(data["G1"], data["G3"])
pyplot.xlabel("G1")
pyplot.ylabel("Final Grade")
```
