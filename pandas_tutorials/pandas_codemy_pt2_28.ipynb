# **PART26**

linear regression

````python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
````

load our diabetes dataset

````python
diabetes = load_diabetes()
print(diabetes)
````

convert data to a pandas dataframe

````python
df6 = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df6
````

add the target

````python
df6['target'] = diabetes.target
df6
````

grab the head

````python
df6.head()
````

# **PART27**

split the data into x and y

````python
X = df6.drop('target', axis=1)
y = df6['target']

X.shape, y.shape
````

````python
from sklearn.model_selection import train_test_split
````

split the data into training and testing

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)
````

print the shape of the shape of the training and testing sets

````python
print('training set: ', X_train.shape, y_train.shape)
print('testing set: ', X_test.shape, y_test.shape)
````

# **PART28**

````python
from sklearn.linear_model import LinearRegression
````

create a linear regression model

````python
lr = LinearRegression()
````

train the model on the training sets

````python
lr.fit(X_train, y_train)
````

predict on the testing set

y_preads are predicted values of the target variable based on the features

once we have those preds we can compare them to the actual values of the target variable
to evaluate the perfomance of the model

````python
y_pred = lr.predict(X_test)
y_pred.shape
````

````python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
````

evaluate performance of the model

````python
# variance of the dependent variable explained by the independent variables (features)
# How well model fits the data. Higher is better
r2 = r2_score(y_test, y_pred)

# Average squared distance between the predicted and actual values. Lower is better
mse = mean_squared_error(y_test, y_pred)

# Average absolute distance between the predicted and actual values. Lower is better
mae = mean_absolute_error(y_test, y_pred)

#starting point of the regression line on the y-axis. value of the dependent variable (target) when the independent variables (features)
#if positive number, target increases as features increase, if negative  - opposite
intercept = lr.intercept_

print('R2 score: ', r2)
print('Mean Squared Error: ', mse)
print('Mean Absolute Error: ', mae)
print('Intercept: ', intercept)
````

````python
import matplotlib.pyplot as plt
````

plot the predicted vs. actual values
predcited values of the traget variable are plotted on the x-axis
actual values of the target are plotted on the y-axis

if the model is a good fit, the points should be close to the diagonal line

indicating a strong linear relationship between the actual and predicted values


````python
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')
plt.xlabel("actual values")
plt.ylabel("predicted values")
plt.title("predcited vs actual values")
plt.show()
````

plot the residuals
the residuals are plotted againts the predicted values of the target variable.

if the linear regression model is a good fit for the data,
the residual plot should show a random scatter of the points around zero
with no discernable trend.



````python
plt.scatter(y_pred, y_test - y_pred, alpha=0.5)
plt.xlabel("predcited values")
plt.ylabel('residuals')
plt.title('residual plot (linear regression)')
````
